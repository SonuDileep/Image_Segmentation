```python
### Visualize a sample image before training###

import pixellib
from pixellib.custom_train import instance_custom_training

vis_img = instance_custom_training()
vis_img.load_dataset("/content/drive/MyDrive/SkySpecs/")
vis_img.visualize_sample()
```

```python
###Training Code###

import pixellib
from pixellib.custom_train import instance_custom_training

train_maskrcnn = instance_custom_training()
train_maskrcnn.modelConfig(network_backbone = "resnet101", num_classes= 1, batch_size = 4)
train_maskrcnn.load_pretrained_model("/content/drive/MyDrive/SkySpecs/mask_rcnn_model.004-0.346451.h5")
train_maskrcnn.load_dataset("/content/drive/MyDrive/SkySpecs")
train_maskrcnn.train_model(num_epochs = 50, augmentation=True,  path_trained_models = "/content/drive/MyDrive/SkySpecs/weights")
#"/content/drive/My Drive/Segmentation"
```

```python
#Testing the model
import pixellib
from pixellib.instance import custom_segmentation

segment_image = custom_segmentation()
segment_image.inferConfig(num_classes= 1, class_names= ["bg","turbine"],detection_threshold = 0.85)
segment_image.load_model("/content/drive/MyDrive/SkySpecs/weights/mask_rcnn_model.032-0.030759.h5") # import the weight

for i in range(119,147):
  img_path = "/content/drive/MyDrive/SkySpecs/sdf/" + str(i) + ".jpg" #input image path
  out_path = "/content/drive/MyDrive/SkySpecs/out/" + str(i) + ".jpg" #output image path
  segment_image.segmentImage(img_path, show_bboxes=True, output_image_name=out_path)
```
